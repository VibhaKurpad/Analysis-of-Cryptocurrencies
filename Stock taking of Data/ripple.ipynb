```python
import pandas as pd
import numpy as np
df = pd.read_csv("ripple_price.csv", parse_dates=['Date'])
```

```python
df_bp = df[['Date','Open', 'Close']]
df_bp = df_bp.set_index('Date')
df_bp['Year'] = df_bp.index.year
df_bp = df_bp.groupby(['Year'])
df1_bp = df_bp.get_group(2013)
df1_bp.boxplot(by='Year')
```
